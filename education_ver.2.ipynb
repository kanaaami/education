## 3-1. ヒストグラム

```python
plt.hist(datarget_new.Pulse, bins=6, rwidth=0.8, range=(45, 75))
```

- bins=6　：ビン（階級数）を6に指定
- rwidth=0.8　：各ビンの幅を0.8に指定
- range=(45, 75)　：ビンの最小値と最大値を45,75に指定

## 3-2. 円グラフ

```python
datarget_Class = datarget_new.groupby('Class').size()
datarget_Class
```

- groupby：特定の列の集計
- size：データの個数


print(type(datarget_Class))を実行するとSeries型であることが分かる．
- Series：1次元のデータを格納できる構造

### 問13. datarget_Classを円グラフで示せ．
- plt.pie(A)
    - AはDataFrame

```python
#解13
plt.pie(datarget_Class)
```

```python
plt.pie(datarget_Class, startangle=90, counterclock=False, colors=['r','g','b','c','m','y'], labels=['A','B','C','D','E','F'])
plt.axis('equal')
```

- startangle=90　：スタート位置を90度左に回転
- counterclock=False　：右回り
- colors=['r','g', ...]　：色をr,g,...と指定
- labels=['A','B', ...]　：ラベルをA,B,...と指定
- plt.axis('equal')　：楕円から円に修正
